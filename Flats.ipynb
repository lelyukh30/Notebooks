```python
def time_to(line):
    if line =='no_bus_stop_nearby':
        return 1000
    minute = line.replace('min', '').replace('-', '~').split('~')
    minute = [int(num) for num in minute]
    return max(minute)
```

```python
def dummies_for(df, col_name):
    dummies = pd.get_dummies(df[col_name])
    for column in dummies.columns:
        df[column] = dummies[column]
    df = df.drop([col_name], axis = 1)
    return df
```

```python
df['HeatingType']=df['HeatingType'].map(lambda x: 1 if x == 'individual_heating' else 0)
df['AptManageType']=df['AptManageType'].map(lambda x: 1 if x =='AptManageType' else 0)
df['TimeToSubway']=df['TimeToSubway'].map(time_to)
df['TimeToBusStop']=df['TimeToBusStop'].map(time_to)
df = dummies_for(df, 'HallwayType')
df = dummies_for(df, 'SubwayStation')
```
